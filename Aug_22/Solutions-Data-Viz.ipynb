## Solution to Data Visualization and EDA the Excercise

```python
#import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
cData = pd.read_csv('countries.csv')
```

```python
cData.info()
cData.head()
```

```python
us = cData[cData.country == 'United States']
india = cData[cData.country == 'India']
```

```python
plt.plot(us.year, us.population/10**6)
plt.plot(india.year, india.population/10**6)
plt.legend(['USA', 'India'])
plt.xlabel('Year')
plt.ylabel('Population (in Millions)')
#plt.show()
```

```python
print(india.population)
print(us.population)
```

```python
print(india.population.iloc[0])
print(us.population.iloc[0])
```

```python
plt.plot(us.year, us.population/us.population.iloc[0] * 100)
plt.plot(india.year, india.population/india.population.iloc[0] *100)
plt.legend(['USA', 'India'])
plt.xlabel('Year')
plt.ylabel('Population Growth %(1st year = 100%)')
#plt.show()
```
